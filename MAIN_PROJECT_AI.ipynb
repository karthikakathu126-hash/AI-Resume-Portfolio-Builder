```python
from transformers import pipeline
from reportlab.lib.pagesizes import A4
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer
from reportlab.lib.styles import getSampleStyleSheet
from reportlab.lib.enums import TA_CENTER
from reportlab.lib.units import inch
```

```python
generator = pipeline(
    "text-generation",
    model="distilgpt2"
)

print("AI Resume Generator Loaded Successfully")
```

```python
name = "KARTHIKA SB"
email = "karthikakathu126@gmail.com"
phone = "+917805298746"
location = "India"

education = "MSc Statistics, KERALA UNIVERSITY (2024–2026)"

skills = [
    "Python",
    "Machine Learning",
    "Data Analysis",
    "Artificial Intelligence"
]

experience = "Completed academic projects in data analysis and machine learning."

career_objective = "To build a career in AI and data science."
```

```python
summary_prompt = f"""
Write a professional resume summary using the following details:
Skills: {', '.join(skills)}
Experience: {experience}
"""

summary = generator(
    summary_prompt,
    max_length=120,
    min_length=80,
    do_sample=True,
    temperature=0.7
)[0]["generated_text"]
```

```python
file_path = "/content/AI_Resume.pdf"

doc = SimpleDocTemplate(file_path, pagesize=A4)
styles = getSampleStyleSheet()
story = []

name_style = styles["Heading1"]
name_style.alignment = TA_CENTER
story.append(Paragraph(name, name_style))
story.append(Spacer(1, 0.2 * inch))

contact_style = styles["Normal"]
contact_style.alignment = TA_CENTER
story.append(Paragraph(f"{email} | {phone} | {location}", contact_style))
story.append(Spacer(1, 0.3 * inch))

story.append(Paragraph("<b>PROFESSIONAL SUMMARY</b>", styles["Heading2"]))
story.append(Paragraph(summary, styles["Normal"]))

doc.build(story)

print("Resume PDF generated successfully")
```
